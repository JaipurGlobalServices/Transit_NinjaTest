```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime, timedelta
import re
import concurrent.futures

# Initialize the WebDriver in headless mode (use the appropriate driver for your browser)
options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(options=options)

# Go to the Amazon product page
driver.get("https://www.amazon.in/dp/B0D6B66GR3")

# Use JavaScript to select the "Most recent" option from the dropdown
try:
    dropdown = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "cm-cr-sort-dropdown"))
    )
    driver.execute_script("arguments[0].value='recent';", dropdown)
    driver.execute_script("arguments[0].dispatchEvent(new Event('change'));", dropdown)
except Exception as e:
    print(f"Error selecting 'Most recent' option: {e}")

# Get current date and the date 10 days ago
today = datetime.now().strftime("%d %B %Y")
ten_days_ago = (datetime.now() - timedelta(10)).strftime("%d %B %Y")

# Define date format patterns
date_patterns = ["%d %B %Y", "%d %b %Y"] # For full month name and abbreviated month name

def parse_review_date(date_str):
    for pattern in date_patterns:
        try:
            return datetime.strptime(date_str, pattern)
        except ValueError:
            continue
    return None

# Extract review elements
try:
    reviews = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located(
            (By.CSS_SELECTOR, "div[data-hook='review']")
        )
    )

    counts = {'review_count': 0}

    def extract_review_details(review, counts):
        try:
            review_date_str = review.find_element(By.CSS_SELECTOR, "span.review-date").text

            # Extract date part from review_date_str
            review_date = parse_review_date(review_date_str.replace('Reviewed in India on ', ''))

            # Check if the review is within the last 10 days
            if review_date and (datetime.now() - review_date).days <= 45:
                counts['review_count'] += 1
                profile_name = (
                    review.find_element(By.CSS_SELECTOR, "span.a-profile-name").text
                    if review.find_elements(By.CSS_SELECTOR, "span.a-profile-name")
                    else "N/A"
                )

                # Extract star rating from class attribute
                try:
                    star_rating_element = review.find_element(
                        By.CSS_SELECTOR, "i.a-icon-star"
                    )
                    star_class = star_rating_element.get_attribute("class")
                    match = re.search(r"a-star-(\d)", star_class)
                    if match:
                        star_rating = match.group(1)  # Extract the numeric value
                    else:
                        star_rating = "N/A"
                except Exception as e:
                    print(f"Error capturing star rating: {e}")
                    star_rating = "N/A"

                # Extract review title by data-hook attribute
                try:
                    review_title_element = review.find_element(By.CSS_SELECTOR, "[data-hook='review-title']")
                    review_title = review_title_element.text if review_title_element else "N/A"
                except Exception as e:
                    print(f"Error extracting review title: {e}")
                    review_title = "N/A"  # If an exception occurs, set as "N/A"

                # Extract review text
                try:
                    review_text = review.find_element(
                        By.CSS_SELECTOR, "span.review-text"
                    ).text
                except Exception as e:
                    print(f"Error extracting review text: {e}")
                    review_text = "N/A"  # If no text is found, set as "N/A"

                # Return extracted details
                return {
                    "Profile Name": profile_name,
                    "Star Rating": star_rating,
                    "Review Title": review_title,
                    "Review Date": review_date.strftime("%d %B %Y"),
                    "Review Text": review_text
                }
        except Exception as e:
            print(f"Error processing review: {e}")
            return None

    # Use parallel processing to extract review details
    with concurrent.futures.ThreadPoolExecutor() as executor:
        results = list(executor.map(lambda review: extract_review_details(review, counts), reviews))

    # Check if there are any reviews from the last 10 days
    if counts['review_count'] == 0:
        print("No reviews available from the last 10 days.")
    else:
        # Print extracted details
        for result in results:
            if result:
                print(f"Profile Name: {result['Profile Name']}")
                print(f"Star Rating: {result['Star Rating']}")
                print(f"Review Title: {result['Review Title']}")
                print(f"Review Date: {result['Review Date']}")
                print(f"Review Text: {result['Review Text']}")
                print("\n" + "-" * 50 + "\n")

except Exception as e:
    print(f"Error loading reviews: {e}")

# Close the driver
driver.quit()
```
